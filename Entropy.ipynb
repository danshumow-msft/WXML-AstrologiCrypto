```python
from kerykeion import AstrologicalSubject
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from scipy.stats import entropy
```

## House Entropy

```python
current_date=datetime.now()
start_date=current_date-timedelta(days=365*85)
current=start_date
end_date=current_date
moonData=[]
jupiterData=[]
mercuryData=[]
sunData=[]
marsData=[]
venusData=[]
saturnData=[]
uranusData=[]
neptuneData=[]
while (current<=end_date):
    for i in range (12):
        year=current.year
        month=current.month
        day=current.day
        subject = AstrologicalSubject("Tejan", year, month, day, i*2, 0, city="Seattle")
        jupiterData.append(subject.jupiter)
        moonData.append(subject.moon)
        mercuryData.append(subject.mercury)
        sunData.append(subject.sun)
        marsData.append(subject.mars)
        venusData.append(subject.venus)
        saturnData.append(subject.saturn)
        uranusData.append(subject.uranus)
        neptuneData.append(subject.neptune)

    current+=timedelta(days=1)
```

```python
sunHouse=[]
moonHouse=[]
mercuryHouse=[]
venusHouse=[]
marsHouse=[]
jupiterHouse=[]
saturnHouse=[]
uranusHouse=[]
neptuneHouse=[]
for dataDict in sunData:
    value=dataDict['house']
    sunHouse.append(value)
for dataDict in moonData:
    value=dataDict['house']
    moonHouse.append(value)
for dataDict in mercuryData:
    value=dataDict['house']
    mercuryHouse.append(value)
for dataDict in venusData:
    value=dataDict['house']
    venusHouse.append(value)
for dataDict in marsData:
    value=dataDict['house']
    marsHouse.append(value)
for dataDict in jupiterData:
    value=dataDict['house']
    jupiterHouse.append(value)
for dataDict in saturnData:
    value=dataDict['house']
    saturnHouse.append(value)
for dataDict in uranusData:
    value=dataDict['house']
    uranusHouse.append(value)
for dataDict in neptuneData:
    value=dataDict['house']
    neptuneHouse.append(value)
```

```python
sunDist={}
moonDist={}
mercuryDist={}
venusDist={}
marsDist={}
jupiterDist={}
saturnDist={}
uranusDist={}
neptuneDist={}
for house in sunHouse:
    if house in sunDist:
        sunDist[house] += 1
    else:
        sunDist[house] = 1
for house in moonHouse:
    if house in moonDist:
        moonDist[house] += 1
    else:
        moonDist[house] = 1
for house in mercuryHouse:
    if house in mercuryDist:
        mercuryDist[house] += 1
    else:
        mercuryDist[house] = 1
for house in venusHouse:
    if house in venusDist:
        venusDist[house] += 1
    else:
        venusDist[house] = 1
for house in marsHouse:
    if house in marsDist:
        marsDist[house] += 1
    else:
        marsDist[house] = 1
for house in jupiterHouse:
    if house in jupiterDist:
        jupiterDist[house] += 1
    else:
        jupiterDist[house] = 1
for house in saturnHouse:
    if house in saturnDist:
        saturnDist[house] += 1
    else:
        saturnDist[house] = 1
for house in uranusHouse:
    if house in uranusDist:
        uranusDist[house] += 1
    else:
        uranusDist[house] = 1
for house in neptuneHouse:
    if house in neptuneDist:
        neptuneDist[house] += 1
    else:
        neptuneDist[house] = 1
```

```python
sunEntropy= entropy(np.array(list(sunDist.values())))
moonEntropy= entropy(np.array(list(moonDist.values())))
mercuryEntropy= entropy(np.array(list(mercuryDist.values())))
venusEntropy= entropy(np.array(list(venusDist.values())))
marsEntropy= entropy(np.array(list(marsDist.values())))
jupiterEntropy= entropy(np.array(list(jupiterDist.values())))
saturnEntropy= entropy(np.array(list(saturnDist.values())))
uranusEntropy= entropy(np.array(list(uranusDist.values())))
neptuneEntropy= entropy(np.array(list(neptuneDist.values())))
```

```python
print ("Sun: " +str(sunEntropy))
print ("Moon: " +str(moonEntropy))
print ("Mercury: " +str(mercuryEntropy))
print ("Venus: " +str(venusEntropy))
print ("Mars: " +str(marsEntropy))
print ("Jupiter: " +str(jupiterEntropy))
print ("Saturn: " +str(saturnEntropy))
print ("Uranus: " +str(uranusEntropy))
print ("Neptune: " +str(neptuneEntropy))
```
