```python
#FUNCTIONS
def entropy(labels, base=None):
  value,counts = np.unique(labels, return_counts=True)
  norm_counts = counts / counts.sum()
  base = 2 if base is None else base
  return -(norm_counts * np.log(norm_counts)/np.log(base)).sum()

def daterange(start, end):
    for n in range(int ((end - start).days)+1):
        yield start + timedelta(n)

def aspect(p1, p2):
  aspect = abs(p1 - p2)
  if aspect > 6:
        aspect = abs(12 - aspect)
  return aspect
```

```python
#COLLECTING DATA OVER TIME
start = date(2000, 1, 1)
end = date(2015, 12, 31)

for date in daterange(start, end):
    for hour in range(12):
      person = AstrologicalSubject("Person", 2000, date.month, date.day, hour * 2, 0, "London", "GB")
      mercuryData.append(translate[person.mercury.house])
      venusData.append(translate[person.venus.house])
      marsData.append(translate[person.mars.house])
      jupiterData.append(translate[person.jupiter.house])
      saturnData.append(translate[person.saturn.house])
      uranusData.append(translate[person.uranus.house])
      neptuneData.append(translate[person.neptune.house])
      plutoData.append(translate[person.pluto.house])
      moonData.append(translate[person.moon.house])
      sunData.append(translate[person.sun.house])
```
