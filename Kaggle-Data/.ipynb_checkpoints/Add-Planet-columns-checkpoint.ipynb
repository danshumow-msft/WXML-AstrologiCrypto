```python
import pandas as pd
from datetime import timedelta, date, datetime
from kerykeion import AstrologicalSubject
pd.options.mode.copy_on_write = True
```

```python
df = pd.read_csv("credit_astro.csv")
```

```python
df
```

```python
def get_entry(row, i):
    string_list =  [int(x) for x in row['Sign_Birth'].strip('[]').split(',')]
    return string_list[i]
```

```python
df['Sun'] = df.apply(lambda row: get_entry(row, 0), axis=1)
df['Moon'] = df.apply(lambda row: get_entry(row, 1), axis=1)
df['Mercury'] = df.apply(lambda row: get_entry(row, 2), axis=1)
df['Venus'] = df.apply(lambda row: get_entry(row, 3), axis=1)
df['Mars'] = df.apply(lambda row: get_entry(row, 4), axis=1)
df['Jupiter'] = df.apply(lambda row: get_entry(row, 5), axis=1)
df['Saturn'] = df.apply(lambda row: get_entry(row, 6), axis=1)
df['Uranus'] = df.apply(lambda row: get_entry(row, 7), axis=1)
df['Neptune'] = df.apply(lambda row: get_entry(row, 8), axis=1)
```

```python
df.drop('Sign_Birth', axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv('credit_astro.csv', index=False)
```
