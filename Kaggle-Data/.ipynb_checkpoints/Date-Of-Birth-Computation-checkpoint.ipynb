```python
translate = {
  "Ari": 1,
  "Tau": 2,
  "Gem": 3,
  "Can": 4,
  "Leo": 5,
  "Vir": 6,
  "Lib": 7,
  "Sco": 8,
  "Sag": 9,
  "Cap": 10,
  "Aqu": 11,
  "Pis": 12
}
```

```python
# use kerykeion to find sign

def add_astrological_info(row):
    subject = AstrologicalSubject("Subject", year = row['Date_Birth'].year, month = row['Date_Birth'].month, day = row['Date_Birth'].day, hour = 12, minute = 17, lng=80, lat=13, city="Chennai", nation="IN")
    astro_data = [
        translate[subject.sun.sign], translate[subject.moon.sign], translate[subject.mercury.sign],
        translate[subject.venus.sign],translate[subject.mars.sign], translate[subject.jupiter.sign],
        translate[subject.saturn.sign], translate[subject.uranus.sign], translate[subject.neptune.sign]
    ]
    return astro_data
```

```python
df['Sign_Birth'] = df.apply(lambda row: add_astrological_info(row), axis=1)
```

```python
df['Encode_Sign_Birth'] = df.apply(lambda row: astro_encoding(row), axis=1)
```

```python
def astro_encoding(row):
    # we want base 12 in each entry
    # for example a list [1,2,4] should map to (4-1)*1 + (2-1)*12 + (1-1)*12^2
    A= row['Sign_Birth']
    N = len(A)
    num = 0
    for i in range(N):
        num += (12**i)*(A[i]-1)
    return num
```
