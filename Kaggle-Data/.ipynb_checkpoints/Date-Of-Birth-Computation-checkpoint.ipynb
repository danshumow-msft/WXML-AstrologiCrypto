```python
import pandas as pd
from datetime import timedelta, date, datetime
pd.options.mode.copy_on_write = True
```

```python
df = pd.read_csv("application_record.csv.zip", compression="zip")
```

```python
df.head()
```

```python
date_posted = pd.to_datetime('2020-03-24')
```

```python
df["Date_Birth"] = date_posted + pd.to_timedelta(df['DAYS_BIRTH'], unit='D')
```

```python
df.head()
```

```python
mask = df['DAYS_EMPLOYED'] <=0
```

```python
df = df[mask]
```

```python
df["Date_Employed"] = date_posted + pd.to_timedelta(df['DAYS_EMPLOYED'], unit='D')
```

```python
df.to_csv("application_record.csv.gz", compression='gzip')
```
