```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeaveOneOut
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('credit_astro.csv')
```

```python
X=df.drop(['Date_Birth','Date_Employed','Sign_Birth', 'AMT_INCOME_TOTAL'], axis=1)
y=df['AMT_INCOME_TOTAL']
rfe = RFE(LinearRegression(), step=1, n_features_to_select=0.4, verbose=1)
rfe.fit(X, y)
rfe_features = X.columns[[r == 1 for r in rfe.ranking_]]
print("Features:", list(rfe_features))
reg = rfe.estimator_
```

```python
X=df[['CODE_GENDER', 'FLAG_OWN_CAR', 'NAME_EDUCATION_TYPE', 'FLAG_OWN_REALTY']]
y=df['AMT_INCOME_TOTAL']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], linestyle='--', color='red')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
