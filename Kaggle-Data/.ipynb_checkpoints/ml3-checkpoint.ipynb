```python
predict = 'FLAG_OWN_CAR'
#predict = 'CODE_GENDER'
#predict='DAYS_BIRTH'
```

```python
Y = credit[predict]
```

```python
X = credit[['CODE_GENDER', 'AMT_INCOME_TOTAL', 'NAME_FAMILY_STATUS', 'CNT_FAM_MEMBERS']]
#X = credit[['FLAG_OWN_CAR', 'AMT_INCOME_TOTAL', 'DAYS_EMPLOYED']]
#X = credit[['FLAG_OWN_REALTY','CNT_CHILDREN','NAME_EDUCATION_TYPE','NAME_HOUSING_TYPE','DAYS_EMPLOYED','CNT_FAM_MEMBERS','higher_Sign_Birth']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=31,stratify=Y)
```

```python
def metrics_score(actual, predicted):
    print(classification_report(actual, predicted))

    cm = confusion_matrix(actual, predicted)
    plt.figure(figsize=(8,5))

    sns.heatmap(cm, annot=True,  fmt='.2f')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
model = DecisionTreeClassifier(random_state=31)
model.fit(X_train, y_train)
```

Fitting with training data

```python
pred_train = model.predict(X_train)
metrics_score(y_train, pred_train)
```

Check performance with test data

```python
pred_test = model.predict(X_test)
metrics_score(y_test, pred_test)
```

Using grid search

```python
# Choose the type of classifier.
estimator = DecisionTreeClassifier(random_state=1)

# Grid of parameters to choose from
parameters = {
    "max_depth": np.arange(4, 12, 2),
    "max_leaf_nodes": np.arange(6, 24, 3),
    "min_samples_split": [10, 30, 50, 70],
}


# Run the grid search
grid_obj = GridSearchCV(estimator, parameters, cv=5,scoring='recall',n_jobs=-1)
grid_obj = grid_obj.fit(X_train, y_train)

# Set the clf to the best combination of parameters
estimator = grid_obj.best_estimator_

# Fit the best algorithm to the data.
estimator.fit(X_train, y_train)
```

It always returns the least max_depth, not sure why yet.
Fitting the best model returned by grid search and fitting with training data.

```python
train_tuned = estimator.predict(X_train)
metrics_score(y_train, train_tuned)
```

```python
pred_tuned = estimator.predict(X_test)
metrics_score(y_test,pred_tuned)
```
