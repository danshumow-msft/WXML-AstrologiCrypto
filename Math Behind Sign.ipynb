```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Planet:
    def __init__ (self,name, orbital_period):
        self.orbital_period=orbital_period
        self.name=name
    def position(self, t):
        return 2*np.pi*(t/self.orbital_period)
```

```python
mercury=Planet('mercury',87.97)
venus = Planet('venus',224.7)
mars=Planet('mars',686.98)
jupiter=Planet('jupiter',4332.82)
saturn=Planet('saturn',10755.70)
uranus=Planet('uranus',30687.15)
neptune=Planet('neptune',60190.03)
planets=[mercury,venus, mars, jupiter, saturn, uranus, neptune]
planet_list=['Mercury', 'Venus', 'Mars','Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

```python
counts = np.zeros((7, 7))
days=1000000
for i, planet in enumerate(planets):
    for j, planet2 in enumerate(planets):
        for t in range(days):
            if (planet.position(t) - planet2.position(t)) < (np.pi / 6.0) and \
                (int(planet.position(t) / (np.pi / 6.0)) == int(planet2.position(t) / (np.pi / 6.0))):
                counts[i][j] += 1.0
counts=counts/days
                
for i in range (7):
    counts[i][i]=0
```

```python
counts*days
```

```python
fig, ax=plt.subplots(figsize=(8,8))
green_cmap = sns.light_palette("#4B2E83", as_cmap=True)
sns.heatmap(counts,cmap=green_cmap, ax=ax, xticklabels=planet_list, yticklabels=planet_list)
```
