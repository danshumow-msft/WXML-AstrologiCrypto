```python
from kerykeion import AstrologicalSubject
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

## Every Two Hours

```python
current_date=datetime.now()
start_date=current_date-timedelta(days=365*45)
current=start_date
end_date=current_date
moonData=[]
jupiterData=[]
mercuryData=[]
neptuneData=[]
while (current<=end_date):
    for i in range (12):
        year=current.year
        month=current.month
        day=current.day
        subject = AstrologicalSubject("Tejan", year, month, day, i*2, 0, city="Seattle")
        jupiterData.append(subject.jupiter)
        moonData.append(subject.moon)
        mercuryData.append(subject.mercury)
        neptuneData.append(subject.neptune)
    current+=timedelta(days=1)
```

```python
jupiterSign=[]
moonSign=[]
mercurySign=[]
neptuneSign=[]
for dataDict in jupiterData:
    value=dataDict['sign']
    jupiterSign.append(value)
for dataDict in moonData:
    value=dataDict['sign']
    moonSign.append(value)
for dataDict in mercuryData:
    value=dataDict['sign']
    mercurySign.append(value)
for dataDict in neptuneData:
    value=dataDict['sign']
    neptuneSign.append(value)
```

```python
sign_distribution_moon={}
for sign in moonSign:
    if sign in sign_distribution_moon:
        sign_distribution_moon[sign] += 1
    else:
        sign_distribution_moon[sign] = 1
```

```python
sign_distribution_jupiter={}
for sign in jupiterSign:
    if sign in sign_distribution_jupiter:
        sign_distribution_jupiter[sign] += 1
    else:
        sign_distribution_jupiter[sign] = 1
```

```python
sign_distribution_mercury={}
for sign in mercurySign:
    if sign in sign_distribution_mercury:
        sign_distribution_mercury[sign] += 1
    else:
        sign_distribution_mercury[sign] = 1
```

```python
sign_distribution_neptune={}
for sign in neptuneSign:
    if sign in sign_distribution_neptune:
        sign_distribution_neptune[sign] += 1
    else:
        sign_distribution_neptune[sign] = 1
```

```python
plt.bar(sign_distribution_mercury.keys(),sign_distribution_mercury.values())
plt.title('Mercury Sign Distribution')
```
