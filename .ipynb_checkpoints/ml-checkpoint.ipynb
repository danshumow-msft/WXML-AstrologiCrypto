```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Importing the Machine Learning models we require from Scikit-Learn
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Importing the other functions we may require from Scikit-Learn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# To get diferent metric scores
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,precision_recall_curve,roc_curve,make_scorer

# Code to ignore warnings from function usage
import warnings;
import numpy as np
warnings.filterwarnings('ignore')
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Number of data points
num_data_points = 10000

# Generate independent data for columns
independent_data_1 = np.random.randn(num_data_points)
independent_data_2 = np.random.randn(num_data_points)

# Generate correlated data for traits and outcome
mean = [0, 0, 0]
covariance = [[1, 0.5, 0.2], [0.5, 1, 0.3], [0.2, 0.3, 1]]  # Adjust covariance as needed
traits_and_outcome_data = np.random.multivariate_normal(mean, covariance, num_data_points)

# Create DataFrame
data = pd.DataFrame({
    'Trait 1': traits_and_outcome_data[:, 0],
    'Trait 2': traits_and_outcome_data[:, 1],
    'Trait 3': traits_and_outcome_data[:, 2],
    'Outcome': traits_and_outcome_data.sum(axis=1),  # Outcome column is sum of traits
    'Independent 1': independent_data_1,
    'Independent 2': independent_data_2
})
```

```python
threshold = data['Outcome'].median()  # You can use any threshold you like

data['Outcome'] = (data['Outcome'] > threshold).astype(int)

print(data.head())
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(data.corr(), annot = True) #annotation shows the numbers in each box
plt.show()
```

From what I'm understanding is that we are trying to show that giving astrological data doesn't give more information by showing that a certain model does equally well with or without the astrological data.

If we use simulated data, or artificial data, we would know their distribution, and whether it is correlated (dependent) with the outcome variable in the beginning. So that doesn't really show what we want to show, because we want to be able to generalize our result to real data.

I think since we don't know the distribution of the real data, we can't just randomly generate data to try to mimic real data. We have to use samples of the real data.

```python
def metrics_score(actual, predicted):
    print(classification_report(actual, predicted))

    cm = confusion_matrix(actual, predicted)
    plt.figure(figsize=(8,5))

    sns.heatmap(cm, annot=True,  fmt='.2f')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
X_corr=data[['Trait 1', 'Trait 2', 'Trait 3']]
X_full = data.drop(columns = 'Outcome')
Y=data['Outcome']
```

```python
X_corr_train,X_corr_test,y_train,y_test=train_test_split(X_corr,Y,test_size=0.30,random_state=1,stratify=Y)
```

```python
model_dt = DecisionTreeClassifier(random_state=1)
model_dt.fit(X_corr_train, y_train)
```

```python
pred_train_dt = model_dt.predict(X_corr_train)
metrics_score(y_train, pred_train_dt)
```

```python
pred_test_dt = model_dt.predict(X_corr_test)
metrics_score(y_test, pred_test_dt)
```

```python
# Choose the type of classifier.
estimator = DecisionTreeClassifier(random_state=1)

# Grid of parameters to choose from
parameters = {
    "max_depth": np.arange(1, 100, 10),
    "max_leaf_nodes": [50, 75, 150, 250],
    "min_samples_split": [10, 30, 50, 70],
}


# Run the grid search
grid_obj = GridSearchCV(estimator, parameters, cv=5,scoring='recall',n_jobs=-1)
grid_obj = grid_obj.fit(X_corr_train, y_train)

# Set the clf to the best combination of parameters
estimator = grid_obj.best_estimator_

# Fit the best algorithm to the data.
estimator.fit(X_corr_train, y_train)
```

```python
dt_tuned = estimator.predict(X_corr_train)
metrics_score(y_train, dt_tuned)
```

```python
y_pred_tuned = estimator.predict(X_corr_test)
metrics_score(y_test,y_pred_tuned)
```

```python
X_full_train,X_full_test,y_train,y_test=train_test_split(X_full,Y,test_size=0.30,random_state=1,stratify=Y)
```

```python
model_dt_full = DecisionTreeClassifier(random_state=1)
model_dt_full.fit(X_full_train, y_train)
```

```python
full_train_dt = model_dt_full.predict(X_full_train)
metrics_score(y_train, full_train_dt)
```

```python
full_test_dt = model_dt_full.predict(X_full_test)
metrics_score(y_test, full_test_dt)
```
